```python
#VIF calculation

import statsmodels.formula.api as sfa

rsq_RnD= sfa.ols('RnD~Admin+Marketing+Profit',data=df).fit().rsquared
vif_RnD= 1/(1-rsq_RnD)
rsq_Admin= sfa.ols('Admin~RnD+Marketing+Profit',data=df).fit().rsquared
vif_Admin= 1/(1-rsq_Admin)
rsq_Marketing= sfa.ols('Marketing~RnD+Admin+Profit',data=df).fit().rsquared
vif_Marketing= 1/(1-rsq_Marketing)

d1= {'Features':['RnD','Admin','Marketing'],'VIF':[vif_RnD,vif_Admin,vif_Marketing]}
vif_dataframe= pd.DataFrame(d1)
vif_dataframe
```

```python
# Modeling

model_1 = sfa.ols('Profit~(Admin+Marketing)',data = df).fit()
model_1.summary()
```

```python
# Q-Q plot to check the distribution

plt.figure(figsize = (10,7))
plt.hist(model_1.resid,bins = 10)

import statsmodels.api as sm
qqplot = sm.qqplot(model_1.resid,line = 'q')
plt.title('QQ plot for residual', c = 'Purple')
```

```python
from statsmodels.graphics.regressionplots import influence_plot

influence_model = model_1.get_influence()
(c,_) = influence_model.cooks_distance
```
