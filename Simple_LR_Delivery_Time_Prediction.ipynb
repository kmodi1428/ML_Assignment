```python
# Modeling

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
import statsmodels.formula.api as sfa
```

```python
model_1= sfa.ols('Delivery_Time~Sorting_Time', data=df_sorted).fit()
model_1.summary()
```

```python
model_2=sfa.ols('Delivery_Time~np.log(Sorting_Time)', data=df_sorted).fit()
model_2.summary()
```

```python
model_3= sfa.ols('Delivery_Time~Sorting_Time+pow(Sorting_Time,2)+np.log(Sorting_Time)', data=df_sorted).fit()
model_3.summary()
```

```python
model_4= sfa.ols('Delivery_Time~np.sqrt(Sorting_Time)',data=df_sorted).fit()
model_4.summary()
```

```python
model_5= sfa.ols('Delivery_Time~np.cbrt(Sorting_Time)',data=df_sorted).fit()
model_5.summary()
```

```python
plt.figure(figsize=(10, 7))
plt.style.use('seaborn-darkgrid')
plt.scatter(df_sorted['Sorting_Time'], df_sorted['Delivery_Time'], color='blue',s=5)
plt.plot(df_sorted['Sorting_Time'], model_3.predict(), color='black')
plt.xlabel('Sorting Time')
plt.ylabel('Delivery Time')
```

```python
#predicted delivery times

predicted = pd.DataFrame()
predicted['Sorting_Time'] = df_sorted['Sorting_Time']
predicted['Delivery_Time'] = df_sorted['Delivery_Time']
predicted['Predicted_Delivery_Time'] = model_3.predict()
predicted

```

```python
actual = [df_sorted['Delivery_Time']]
predict = [predicted['Predicted_Delivery_Time']]
error= [x-y for x, y in zip(actual, predict)]
error
```

```python
#here we can see that there is no correlaltion between the error and the variable

plt.scatter(error,df_sorted['Sorting_Time'],c = 'red')
plt.xlabel('Sorting_Time',c='black')
plt.ylabel('error',c='black')

```
